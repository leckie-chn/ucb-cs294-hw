```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
exp_names = os.listdir('./data/')
result_avg = {}
result_max = {}
for exp_name in exp_names:
    avg_rets = []
    max_rets = []
    for seed in os.listdir(os.path.join('./data', exp_name)):
        df_log = pd.read_csv(os.path.join('./data/', exp_name, seed, 'log.txt'), sep='\t')
        avg_rets.append(df_log['AverageReturn'].values)
        max_rets.append(df_log['MaxReturn'].values)
    result_avg[exp_name] = np.mean(avg_rets, axis=0)
    result_max[exp_name] = np.mean(max_rets, axis=0)
```

```python
df_avg = pd.DataFrame(result_avg)
df_avg.plot(kind='line', figsize=(15, 8))
plt.show()
```

```python
df_max = pd.DataFrame(result_max)
df_max.plot(kind='line', figsize=(15, 8))
plt.show()
```
